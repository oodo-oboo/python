```python
import pandas as pd
```

```python
df = pd.read_csv('age.csv',encoding='cp949', index_col=0)
df.head()
```

```python
df = df.dropna()
```

```python
df = df.replace(',','',regex=True)
```

```python
a = df.astype('int')
```

```python
# 인구수를 총 인구수로 나눈다 , 인구수 > 비율
b = a.div(a.iloc[:,0], axis=0) # 비율 데이터
c= b.iloc[:,2:] # 컬럼 2개 빼고 선택
c.head(1)
```

```python
c['서울특별시']
```

```python
# 궁금한 지역의 이름을 입력받는다.
name = input('궁금한 지역이 어디에요? : ')
name
```

```python
c.index.str.contains(name).sum()
mask = c.index.str.contains(name)
df2 = c[mask]
df2.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family = 'Malgun Gothic')
df2.T.plot()
```

```python
a.dtypes
```
