```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0,10,(100,2)).astype(np.float32)
x.shape
```

```python
plt.scatter(x[:,0], x[:,1])
plt.xlim(-20, 20)
plt.ylim(-20, 20)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, Normalizer
s0 = MinMaxScaler().fit(x)
x0  = s0.transform(x)

s1 = StandardScaler().fit(x)
x1 = s1.transform(x)

s2 = RobustScaler().fit(x)
x2 = s2.transform(x)

s3 = Normalizer().fit(x)
x3 = s3.transform(x)
```

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(x0[:,0], x0[:,1])
plt.scatter(x1[:,0], x1[:,1])
plt.scatter(x2[:,0], x2[:,1])
plt.scatter(x3[:,0], x3[:,1])
#plt.xlim(-10, 10)
#plt.ylim(-10, 10)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# 유방암 데이터를 2차원으로 주성분을 뽑아서 그래프로 표현합니다.
cancer = load_breast_cancer()
cancer.data.shape # 30차원
```

```python
scaler = StandardScaler().fit(cancer.data)
```

```python
x = scaler .transform(cancer.data)
np.max(cancer.data), np.max(x), x.shape
```

```python
pca = PCA(2).fit(x)
```

```python
x_pca = pca.transform(x)
x_pca.shape # 30차원을 2차원으로
```
