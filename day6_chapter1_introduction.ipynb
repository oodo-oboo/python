- 총 150 개의 데이터밖에 없습니다.

```python
import numpy as np
#np.random.shuffle(iris_dataset.target)
# iris_dataset.target
```

```python
#  데이터를 학습용 75%, 테스트용 25% 로 나눠서 섞습니다.
# x = iris_dataset_data, y = iris_dataset.target
# x_train, x_test, y_train, y_test = train_test_split(x,y)
# random_state=0 이걸 넣으면 고정을 시켜서 난수 값이 변경되지 않도록 set (재현위해)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# p.45  산점도 scatter 
iris_dataframe = pd.DataFrame(x_train, columns= iris_dataset.feature_names)
iris_dataframe.head()
```

```python
plt.rc('font', family = 'Malgun gothic')
df = plt.plot(iris_dataframe)
```

```python
p = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker='o',
                              hist_kwds={'bins':20}, s =60, alpha =.8, cmap = mglearn.cm3)
```

```python
# p.47 최근접 이웃 모임
from sklearn.neighbors import KNeighborsClassifier # 대문자로 시작하면 클래스
```

```python
 # 객체생성, n_neighbors=1이건 하단의 array의 1번째 값과 유사한 값을 찾겠다
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# p.48  훈련 데이터셋으로 모델을 만드는 fit 함수
knn.fit(x_train, y_train) # fit  모델을 훈련 데이터에 맞게 피팅시켜라
```

```python
plt.title('iris_dataset.feature_names[0]')
plt.scatter(np.arange(len(x_train)), x_train[:,0], c=y_train, cmap=mglearn.cm3)
```

```python
# p.48 예측하기
x_new = np.array([[5,2.9,1, 0.2]]) # 야생에서 꽃을 발견했다.
x_new.shape
```

```python
# p.49 
prediction = knn.predict(x_new) #예측해라

print('예측 : ', prediction)
print('예측한 타겟의 이름은 ', iris_dataset.target_names[prediction])
```

```python
# p. 49 모델 평가하기
y_pred = knn.predict(x_test)
print('테스트 세트에 대한 예측값 ')
print('예측 : ' , y_pred)
print('정답 : ' , y_test)
```

```python
print('정확도', knn.score(x_test, y_test))
```
